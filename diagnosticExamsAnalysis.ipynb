```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import scipy.stats as stats
import distribution as dist
```

```python
dioExams = pd.read_csv("DiagnosticExams.csv")
dioExams.head()
```

```python
dioExams = dioExams[dioExams["Measure"] == "Per 1 000 population"]

dioExams.head()
```

```python
wantedData = dioExams[["Country", "Year", "Percentage"]]
wantedData.head()
```

```python
xCol = "Percentage"
valueMean, valueStd = dist.dist_info(wantedData, xCol)
dist.dist_plot(wantedData, xCol, valueMean, valueStd, 0, 200)
wantedData = dist.add_norm(wantedData, xCol, "Percentage")
wantedData.head()
```
